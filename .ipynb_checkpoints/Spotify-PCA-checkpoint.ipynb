```python
# PCA przeprowadzone na bazie 20 krajów z różnych kontynentów. Gdzie cechami każdego kraju jest jego położenie geograficzne, 
# ludność, trzy najbardziej popularne gatunki muzyczne, poziomy parametrów opisujących utwory muzyczne z top100 najpopluarniejszych
# takie jak: danceability, energy, acousticness, instrumentalness, liveness, tempo.
```

```python
import pandas as pd 
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_boston
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
import os

%matplotlib inline
```

```python
df_countries = pd.read_csv(os.getcwd()+'\data\country_data_pca.csv')
```

```python
df_countries
```

```python
sns.pairplot(df_countries, hue='Region')
```

```python
y_region = df_countries.loc[:,'Region'].values
```

```python
y_region
```

```python
x_features_numerical = df_countries.loc[:, ['Danceability','Energy', 'Acousticness', 'Instrumentalness', 'Liveness', 'Tempo']].values
x_features_numerical
```

```python
pca_countries_regions = PCA(n_components=2)
```

```python
X_pca = pca_countries_regions.fit_transform(x_features_numerical)
```

```python
principal_df_regions = pd.DataFrame(data = X_pca ,columns = ['principal component 1', 'principal component 2'])
```

```python
principal_df_regions = pd.concat([principal_df_regions, df_countries[['Region','Country']]], axis = 1)
principal_df_regions
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA - without music genres', fontsize = 20)


region_targets = ['East_EU', 'WEST_EU', 'NORTH_AMERICA', 'SOUTH_AMERICA', 'ASIA', 'SCANDINAVIA', 'OCEANIA']
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
for target, color in zip(region_targets,colors):
    indicesToKeep = principal_df_regions['Region'] == target
    ax.scatter(principal_df_regions.loc[indicesToKeep, 'principal component 1']
               , principal_df_regions.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
for row in principal_df_regions.values:
    ax.annotate(row[3], (row[0],row[1]))
ax.legend(region_targets)
ax.grid()
```

```python
x_genres_strings = df_countries.loc[:, ['Top 1 genre','Top 2 genre','Top 3 genre']].values
x_genres_strings
```

```python
genres_dict = {}
genres_code = 0
for country_genres in x_genres_strings:
    for genre in country_genres:
        if genre not in genres_dict:
            genres_dict[genre] = genres_code
            genres_code += 1
```

```python
genres_dict
```

```python
x_genres_numerical = []
for country_genres in x_genres_strings:
    country_genres_numerical = []
    for genre in country_genres:
        country_genres_numerical.append(genres_dict[genre])
    x_genres_numerical.append(country_genres_numerical)
x_genres_numerical = np.array(x_genres_numerical)
```

```python
x_genres_numerical
```

```python
df_countries['Top 1 genre num'] = x_genres_numerical[:,0]
df_countries['Top 2 genre num'] = x_genres_numerical[:,1]
df_countries['Top 3 genre num'] = x_genres_numerical[:,2]
```

```python
df_countries
```

```python
sns.pairplot(df_countries, hue='Region')
```

```python
x_features_numerical_with_genres = df_countries.loc[:, ['Danceability','Energy', 'Acousticness', 'Instrumentalness', 'Liveness', 'Tempo', 'Top 1 genre num', 'Top 2 genre num', 'Top 3 genre num']].values
x_features_numerical_with_genres
```

```python
pca_countries_regions = PCA(n_components=2)
```

```python
X_pca = pca_countries_regions.fit_transform(x_features_numerical_with_genres)
```

```python
principal_df_regions = pd.DataFrame(data = X_pca ,columns = ['principal component 1', 'principal component 2'])
```

```python
principal_df_regions = pd.concat([principal_df_regions, df_countries[['Region','Country']]], axis = 1)
principal_df_regions
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA - with music genres', fontsize = 20)


region_targets = ['East_EU', 'WEST_EU', 'NORTH_AMERICA', 'SOUTH_AMERICA', 'ASIA', 'SCANDINAVIA', 'OCEANIA']
colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
for target, color in zip(region_targets,colors):
    indicesToKeep = principal_df_regions['Region'] == target
    ax.scatter(principal_df_regions.loc[indicesToKeep, 'principal component 1']
               , principal_df_regions.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
for row in principal_df_regions.values:
    ax.annotate(row[3], (row[0],row[1]))
ax.legend(region_targets)
ax.grid()
```

```python
genres_dict_fixed = {}
genres_code = 0
for country_genres in x_genres_strings:
    for genre in country_genres:
        if 'hip hop' in genre:
            genre = 'hip hop'
        if 'trap' in genre:
            genre = 'trap'
        if 'rap' in genre:
            genre = 'rap'
        if 'pop' in genre and not genre == 'dance pop':
            genre = 'pop'
        if genre not in genres_dict_fixed:
            genres_dict_fixed[genre] = genres_code
            genres_code += 1
```

```python
genres_dict_fixed
```

```python
x_genres_numerical_fixed = []
for country_genres in x_genres_strings:
    country_genres_numerical = []
    for genre in country_genres:
        if 'hip hop' in genre:
            genre = 'hip hop'
        if 'trap' in genre:
            genre = 'trap'
        if 'rap' in genre:
            genre = 'rap'
        if 'pop' in genre and not genre == 'dance pop':
            genre = 'pop'
        country_genres_numerical.append(genres_dict_fixed[genre])
    x_genres_numerical_fixed.append(country_genres_numerical)
x_genres_numerical_fixed = np.array(x_genres_numerical_fixed)
```

```python
x_genres_numerical_fixed
```

```python
df_countries['Top 1 genre num'] = x_genres_numerical_fixed[:,0]
df_countries['Top 2 genre num'] = x_genres_numerical_fixed[:,1]
df_countries['Top 3 genre num'] = x_genres_numerical_fixed[:,2]
```

```python
x_features_numerical_with_genres = df_countries.loc[:, ['Danceability','Energy', 'Acousticness', 'Instrumentalness', 'Liveness', 'Tempo', 'Top 1 genre num', 'Top 2 genre num', 'Top 3 genre num']].values
x_features_numerical_with_genres
```

```python
pca_countries_regions = PCA(n_components=2)
```

```python
X_pca = pca_countries_regions.fit_transform(x_features_numerical_with_genres)
```

```python
principal_df_regions = pd.DataFrame(data = X_pca ,columns = ['principal component 1', 'principal component 2'])
```

```python
principal_df_regions = pd.concat([principal_df_regions, df_countries[['Region','Country']]], axis = 1)
principal_df_regions
```
